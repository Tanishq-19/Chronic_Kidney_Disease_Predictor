# Chronic Kidney Disease Predictor

This notebook uses Machine learning models to predict the chronic kidney disease using python-based machine learning and data science libraries.

## 1. Problem Definition

Problem is to explore if a patient has chronic kidney disease or not!

## 2. Data

The data was taken over a 2-month period in India with 25 features ( eg, red blood cell count, white blood cell count, etc). The target is the 'classification', which is either 'ckd' or 'notckd' - ckd=chronic kidney disease. Use machine learning techniques to predict if a patient is suffering from a chronic kidney disease or not.
Original dataset can be acquired from the link Chronic KIdney Disease (https://www.kaggle.com/mansoordaku/ckdisease)

## 3. Evaluation
The goal is to train model to provide more that 90% or more than that of accuracy.

### Filling the missing values

```python
for label, content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
# Filling the numeric missing values with median of the feature values
for label, content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            # Adding binary column to notify whether this data was missing 
            df[label+" is_missing"] = pd.isnull(content)
            # Filling missing numeric value with the median
            df[label] = content.fillna(content.median())
```

```python
df.isna().sum()
```

```python
# Filling the categorical data by converting them into numbers

# First checking the categorical values
for label, content in df.items():
    if not pd.api.types.is_numeric_dtype(content):
        print(label)
```

```python
# Filling the categorical data by converting them into numbers

for label, content in df.items():
    if not pd.api.types.is_numeric_dtype(content):
        # Adding binary column to notify whether data was null 
        df[label+" is_missing"] = pd.isnull(content)
        # Turning category into number and adding 1
        df[label] = pd.Categorical(content).codes+1
```

## Modeling

```python
# Importing modeling libraries 
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
# Getting the datasets ready
X = df_temp.drop("classification", axis=1)
y = df_temp.classification 

# Seeding value
seed_value = 42

np.random.seed(seed_value)

# Spliting the datasets into train and test datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Dictionary of models
models = {"SVC": SVC(),
          "Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "RandomForest Classifier": RandomForestClassifier()
         }

# Function to fit and score the models
def fit_and_score(models, X_train, X_test, y_train, y_test):
    """
     Fit and scores the model from the models dictionary.
     X_train: Training dataset (no labels)
     X_test: Testing dataset (np labels)
     y_train: Training labels
     y_test: Training labels
    """
    
    #  Setting random seed
    np.random.seed(seed_value)
    
    # Empty dictionary for model scores
    model_scores= {}
    
    # Looping through all models to fit and score
    for name, model in  models.items():
        # Fitting model
        model.fit(X_train, y_train)
        
        # Scoring and addding the values to model_scores dict
        model_scores[name] = model.score(X_test, y_test)
        
    return model_scores
        
```

```python
# Using the fit_and_score function
model_scores = fit_and_score(models, X_train, X_test, y_train, y_test)
model_scores
```

Therefore the RandomForestClassifier is best with given dataset

```python
# Initializing and fitting the model on RandomForestCLassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)

model.score(X_test, y_test)
```

#### Confusion Matrix

```python
# Importing confusion_matrix
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_preds, y_test))
```

```python
# Defining function to plot confusion matrix
def plot_confusion_matrix(y_preds, y_test):
    """
    Plots the confusion matrix using python's seaborn library.
    """
    fig, ax = plt.subplots(figsize=(4, 4))
    
    ax = sbs.heatmap(confusion_matrix(y_preds, y_test),
                     annot = True,
                     cbar=False)
    plt.xlabel("True label")
    plt.ylabel("Predicted label")
    
plot_confusion_matrix(y_preds, y_test)
```

#### Classification Report

```python
# importing Classifier report
from sklearn.metrics import classification_report

print(classification_report(y_preds, y_test))
```

#### Evaluation Matrix using Cross-Validation

```python
# Importing cross_val_score for calculating accuracy, precision, recall and f1 score
from sklearn.model_selection import cross_val_score
```

```python
# Cross-Validated Accuracy
cv_acc = cross_val_score(model, X, y, cv=5, scoring="accuracy")
cv_acc = np.mean(cv_acc)
cv_acc
```

```python
# Cross-Validated Precision
cv_prec = cross_val_score(model, X, y, cv=5, scoring="precision")
cv_prec = np.mean(cv_prec)
cv_prec
```

```python
# Cross-Validated Recall
cv_rec = cross_val_score(model, X, y, cv=5, scoring="recall")
cv_rec = np.mean(cv_rec)
cv_rec
```

```python
# Cross-Validated f1 score
cv_f1 = cross_val_score(model, X, y, cv=5, scoring="f1")
cv_f1 = np.mean(cv_f1)
cv_f1
```

```python
# Visualize cross-validated metrics
cv_metrics = pd.DataFrame({"Accuracy":cv_acc,
                            "Precision": cv_prec,
                            "Recall": cv_rec,
                            "f1": cv_f1},
                            index=[0])

cv_metrics.T.plot.bar(title="Cross-Validated Classification Metrics", 
                      legend=False);
```

### Feature Importance

```python
# Finding the feature importances
model.feature_importances_
```

```python
# Helper function for plotting feature importance
def plot_features(columns, importances, n=20):
    df = (pd.DataFrame({"features": columns,
                        "feature_importances": importances})
          .sort_values("feature_importances", ascending=False)
          .reset_index(drop=True))
    
    # Plot the dataframe
    fig, ax = plt.subplots()
    ax.barh(df["features"][:n], df["feature_importances"][:n])
    ax.set_ylabel("Features")
    ax.set_xlabel("Feature importance")
    ax.invert_yaxis()
```

```python
# Ploting features 
plot_features(X_train.columns, model.feature_importances_)
```
